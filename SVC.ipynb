```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = load_iris()
```

```python
X = iris.data[:,0:4:3]
y = iris.target
plt.scatter(X.T[0],X.T[1], c=y)
```

```python
from sklearn import svm
from sklearn.preprocessing import scale
```

```python
svml = svm.SVC(kernel='poly', degree=4)
```

```python
svml.fit(X,y)
```

```python
xx,yy = np.meshgrid(np.linspace(0.0,10.0,201),np.linspace(0.0,10.0,201))

pred_svm = svml.predict(np.c_[xx.ravel(),yy.ravel()])
Z_svml = pred_svm.reshape(np.shape(xx))
```

```python
plt.figure(figsize=(14,7))

plt.subplot(121)
plt.contourf(xx,yy,Z_svml,alpha=0.4)
plt.scatter(X.T[0],X.T[1],c=y)
plt.axis('scaled')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('SVC',fontsize=20)
plt.xlim(2.0,10.0)
plt.ylim(0.0,8.0)
```
