```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_iris
iris = load_iris()
```

```python
data = iris.data
target = iris.target
```

```python
iris.feature_names
```

```python
plt.scatter(data[:,1], data[:,2], cmap='brg')
```

## KMeans

```python
from sklearn.cluster import KMeans
kms = KMeans(n_clusters = 5)
kms.fit(data[:,1:3:1])
y_pred = kms.predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1], data[:,2], c=y_pred, cmap='brg')
```

```python
kms.predict([[3.0, 5.0],[4.0, 6.0]])
```

## MeanShift

```python
from sklearn.cluster import MeanShift
msh = MeanShift(bandwidth = 0.9) #h value...small h correspond to many clusters and big h to few clusters
msh.fit(data[:,1:3:1])
y_pred = msh.predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1], data[:,2], c=y_pred, cmap='brg')
```

## Affinity Propagation

```python
from sklearn.cluster import AffinityPropagation
afp = AffinityPropagation() 
```

```python
afp.fit(data[:,1:3:1])
y_pred = afp.predict(data[:,1:3:1])
```

```python
plt.scatter(data[:,1], data[:,2], c=y_pred, cmap='brg')
```

```python
np.unique(afp.labels_).size #The Number of Clusters
```

```python
from sklearn.cluster import AffinityPropagation
afp = AffinityPropagation(damping=0.9) #damping has effect on the number of clusters
afp.fit(data[:,1:3:1])
y_pred = afp.predict(data[:,1:3:1])

plt.scatter(data[:,1], data[:,2], c=y_pred, cmap='brg')
```

```python
np.unique(afp.labels_).size #The Number of Clusters
```

## Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
agc = AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='complete') 
# agc.fit(data[:,1:3:1])
y_pred = agc.fit_predict(data[:,1:3:1])

plt.scatter(data[:,1], data[:,2], c=y_pred, cmap='brg')
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
dbs = DBSCAN(eps=0.5, min_samples=5) #min_samples means Number of neighbours
y_pred = dbs.fit_predict(data[:,1:3:1])

plt.scatter(data[:,1], data[:,2], c=y_pred, cmap='brg')
```
